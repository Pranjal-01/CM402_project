```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 4))

# Plot data in the first subplot (ax1)
ax1.imshow(np.flip(R[0],axis=0))
ax1.set_title('Year 1950 (Root-Shoot Ratio)')

# Plot data in the second subplot (ax2)
ax2.imshow(np.flip(R[11],axis=0))
ax2.set_title('Year 2005 (Root-Shoot Ratio)')

# Adjust the layout and show the plots
plt.tight_layout()
plt.show()
```

```python
# Create a figure with subplots
fig, axs = plt.subplots(5, 3, figsize=(15, 15))
fig.suptitle("Forest Carbon Stock")

# Loop through the first dimension (i)
for i in range(13):
    row, col = i //3 , i % 3  # Determine the subplot position

    # Plot the data for the current i
    axs[row, col].imshow(np.flip(fstockch[(i), :, :],axis=0), cmap='viridis')  # You can choose a different colormap
    axs[row, col].set_title(f'year {1950+i*5}')

# Remove empty subplots, if any
for i in range(13, 16):
    fig.delaxes(axs[i // 3, i % 3])

# Adjust the layout and show the plots
plt.tight_layout()
plt.show()
```
